```python
# one hot encoding
bournemouth_onehot = pd.get_dummies(df[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
bournemouth_onehot['Venue Name'] = df['Venue Name'] 

# move neighborhood column to the first column
fixed_columns = [bournemouth_onehot.columns[-1]] + list(bournemouth_onehot.columns[:-1])
bournemouth_onehot = bournemouth_onehot[fixed_columns]

bournemouth_onehot
```

```python
from sklearn.cluster import KMeans
```

```python
kclusters = 4

bournemouth_grouped_clustering = bournemouth_onehot.drop('Venue Name', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(bournemouth_grouped_clustering)

kmeans.labels_
```

```python
bournemouth_grouped_clustering.insert(0, 'Cluster Labels', kmeans.labels_)
```
